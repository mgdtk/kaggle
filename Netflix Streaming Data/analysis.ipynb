## Analyzing Data

### Analyzing the number of titles available on Netflix by year of release

A large portion of the available content was released in the 21st century, with a noticeable decrease in releases during the pandemic.

```python
release_count = df['release_year'].value_counts().reset_index()

plt.figure(figsize=(12,6))

sns.barplot(data=release_count, x='release_year', y='count', hue='release_year', palette='viridis', legend=False)

plt.title('Number of Titles Available on Netflix by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.show()
```

### Visualizing the proportion of Movies vs TV Shows available on Netflix

The majority of the available titles are movies.

```python
plt.figure(figsize=(6,5))

df['type'].value_counts().plot(kind='bar', color=['lightblue', 'salmon'])

plt.title('Distribution of Content Types on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Number of Titles')
plt.xticks(rotation=0)

plt.show()
```

### Examining Netflix's content by country of production

Most of the available titles originate from countries with English as the main language.

```python
plt.figure(figsize=(16,6))
plt.suptitle('Distribution of Netflix Content by Country', fontsize=16, fontweight='bold', y=1.02)

plt.subplot(1, 2, 1)

movie_countries = (
    df[(df['type'] == 'Movie') & (~df['country'].isnull())]
    ['country']
    .value_counts()
    .head(10)
    .reset_index()
)

sns.barplot(data=movie_countries, x='country', y='count', hue='country', palette='flare')
plt.title('Top 10 Countries With the Most Movies on Netflix', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)

tvshow_countries = (
    df[(df['type'] == 'TV Show') & (~df['country'].isnull())]
    ['country']
    .value_counts()
    .head(10)
    .reset_index()
)

sns.barplot(data=tvshow_countries, x='country', y='count', hue='country', palette='flare')
plt.title('Top 10 Countries With the Most TV Shows on Netflix', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of TV Shows', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.show()
```

### Standardizing and visualizing Netflix's content ratings

It is evident that a large portion of the content available is directed toward adult viewers.

```python
df['rating'] = df['rating'].replace('UR', 'NR')
df['rating'] = df['rating'].replace('NC-17', 'R')
df['rating'] = df['rating'].replace('TV-Y7-FV', 'TV-Y7')

data = df['rating'].value_counts().reset_index()
data = data[data['count'] > 1]

plt.figure(figsize=(8,5))

sns.barplot(data=data, x='rating', y='count', hue='rating', palette='mako_r')
plt.title('Distribution of Netflix Content by Rating')
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45)
plt.show()
```
